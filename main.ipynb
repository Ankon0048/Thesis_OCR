```python
# Update your base directories
source_folder = r"D://Thesis//Mass_Line_Extraction//output_images_line//printed"
destination_folder = r"D://Thesis//Mass_Line_Extraction//data//printed"
```

```python
# Create the destination folder if it doesn't exist
os.makedirs(destination_folder, exist_ok=True)

# Define allowed image extensions (you can modify as needed)
image_extensions = {'.jpg', '.jpeg', '.png', '.bmp', '.gif', '.tiff'}

# Initialize a counter to handle duplicate filenames
counter = 1

# Walk through all subdirectories and files in the source folder
for root, dirs, files in os.walk(source_folder):
    for file in files:
        if os.path.splitext(file)[1].lower() in image_extensions:
            source_path = os.path.join(root, file)
            
            # Handle duplicate filenames by renaming
            filename = file
            dest_path = os.path.join(destination_folder, filename)
            while os.path.exists(dest_path):
                name, ext = os.path.splitext(file)
                filename = f"{name}_{counter}{ext}"
                dest_path = os.path.join(destination_folder, filename)
                counter += 1
            
            # Copy image
            shutil.copy2(source_path, dest_path)
            print(f"Copied: {source_path} -> {dest_path}")

print("All images have been copied.")
```

```python
import os
import cv2
import albumentations as A
from tqdm import tqdm

# Strong augmentation pipeline
augment = A.Compose([
    A.RandomRotate90(),
    A.HorizontalFlip(p=0.5),
    A.VerticalFlip(p=0.2),
    A.MotionBlur(p=0.2),
    A.GaussianBlur(p=0.2),
    A.GaussNoise(var_limit=(10.0, 50.0), p=0.3),
    A.RandomBrightnessContrast(p=0.3),
    A.HueSaturationValue(p=0.3),
    A.Perspective(scale=(0.05, 0.1), p=0.3),
    A.GridDistortion(p=0.3),
    A.ElasticTransform(alpha=1, sigma=50, alpha_affine=50, p=0.2),
])

# Input and output folders
input_dir = r"D://Thesis//Mass_Line_Extraction//data"  # update path
output_dir = r"D://Thesis//Mass_Line_Extraction//Augmentated_images"  # update path

def process_and_augment_images(input_path, output_path):
    if not os.path.exists(output_path):
        os.makedirs(output_path)

    for root, _, files in os.walk(input_path):
        for file in tqdm(files, desc=f"Processing {root}"):
            if not file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
                continue

            img_path = os.path.join(root, file)
            img = cv2.imread(img_path)
            if img is None:
                continue

            augmented = augment(image=img)["image"]

            # Create relative path for saving
            rel_path = os.path.relpath(root, input_path)
            save_dir = os.path.join(output_path, rel_path)
            os.makedirs(save_dir, exist_ok=True)

            save_path = os.path.join(save_dir, f"aug_{file}")
            cv2.imwrite(save_path, augmented)

if __name__ == "__main__":
    process_and_augment_images(input_dir, output_dir)
    print("Augmentation completed.")
```

```python
import os
import shutil
import random
from collections import defaultdict

def split_dataset_by_group(
    src_root, 
    dst_root, 
    train_ratio=0.8, 
    seed=42
):
    random.seed(seed)
    
    classes = ['handwritten', 'printed']
    
    for cls in classes:
        src_dir = os.path.join(src_root, cls)
        dst_train_dir = os.path.join(dst_root, 'train', cls)
        dst_val_dir = os.path.join(dst_root, 'val', cls)

        os.makedirs(dst_train_dir, exist_ok=True)
        os.makedirs(dst_val_dir, exist_ok=True)

        # Group images by base name (original + augmentations)
        groups = defaultdict(list)
        for fname in os.listdir(src_dir):
            if fname.lower().endswith(('.png', '.jpg', '.jpeg')):
                base = fname.split('_aug')[0].split('.')[0]  # e.g., 'img001'
                groups[base].append(fname)

        # Shuffle and split
        keys = list(groups.keys())
        random.shuffle(keys)
        train_keys = keys[:int(len(keys) * train_ratio)]
        val_keys = keys[int(len(keys) * train_ratio):]

        for key in train_keys:
            for fname in groups[key]:
                shutil.copy(os.path.join(src_dir, fname), os.path.join(dst_train_dir, fname))

        for key in val_keys:
            for fname in groups[key]:
                shutil.copy(os.path.join(src_dir, fname), os.path.join(dst_val_dir, fname))

        print(f"Class '{cls}': {len(train_keys)} groups for training, {len(val_keys)} for validation.")

# Example usage
src_dataset_path = "D://Thesis//Mass_Line_Extraction//data//data"  # change this
dst_dataset_path = "D://Thesis//Mass_Line_Extraction//dataset_with_different_val_train"

split_dataset_by_group(src_dataset_path, dst_dataset_path)
```

```python
import torch
import torch.nn as nn
from torchvision import transforms
from PIL import Image

# Load the entire model (assuming you used torch.save(model, "model.pth"))
model = torch.load("D://Thesis//Mass_Line_Extraction//test//final_model.pth", weights_only=False)
model.eval()

# If using GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)
```

```python
IMG_SIZE = 128  # must match training

transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5], std=[0.5])  # For grayscale
])
```

```python
# Path to your test image
image_path = "D://Thesis//Mass_Line_Extraction//test//4VXPUZNQ7BDHEE6NUIUR4GN4AU.png"

# Load and preprocess
image = Image.open(image_path).convert("L")  # Convert to grayscale
input_tensor = transform(image).unsqueeze(0).to(device)  # Add batch dimension

# Predict
with torch.no_grad():
    output = model(input_tensor)
    predicted_class = torch.argmax(output, dim=1).item()

# Map to label names
class_names = ['handwritten', 'printed']  # Update if your class order is different
print(f"Predicted class: {class_names[predicted_class]}")
```
